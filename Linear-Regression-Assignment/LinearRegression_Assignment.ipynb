## Step 4. Data Cleanup

```python
# Cleanup Data
# Creating Derived Variable 'carcompany' from 'CarName' field as we only need company name 
# as the independent variable for model building 
def mapCarCompany(x):
    return x.split()[0]
cardataset['CarName']=cardataset['CarName'].str.replace('-',' ')
cardataset['CarName']=cardataset['CarName'].apply(mapCarCompany)
cardataset.CarName.unique()
```

```python
# There are some redundant data in carName.lets fix those by cleaning.
cardataset.CarName = cardataset['CarName'].str.lower()
cardataset['CarName'] = cardataset['CarName'].str.replace('maxda','mazda')
cardataset['CarName'] = cardataset['CarName'].str.replace('porcshce','porsche')
cardataset['CarName'] = cardataset['CarName'].str.replace('toyouta','toyota')
cardataset['CarName'] = cardataset['CarName'].str.replace('vokswagen','volkswagen')
cardataset['CarName'] = cardataset['CarName'].str.replace('vw','volkswagen')
cardataset.CarName.unique()
```

## Step 5. Categorical Field treatment

```python
# Changing Categorical Data to numerical/binary level
cardataset['fueltype'] = cardataset['fueltype'].map({'gas': 1, 'diesel': 0})
cardataset['aspiration'] = cardataset['aspiration'].map({'std': 1, 'turbo': 0})
cardataset['doornumber'] = cardataset['doornumber'].map({'two': 1, 'four': 0})
cardataset['enginelocation'] = cardataset['enginelocation'].map({'front': 1, 'rear': 0})
```

```python
cardataset.info()
```

```python
# Dummy values for categorical columns of more than 2 levels
df = pd.get_dummies(cardataset)
df.head()
```

## Step 6. Rescale Features by Normalisation

```python
# Rescale Features by Normalisation by defining a normalisation function (only numeric fields)
columns_to_normalize = ['carlength','carheight','symboling','boreratio','enginesize','stroke','compressionratio','peakrpm','horsepower','citympg','price']
normalised_df = df[columns_to_normalize].apply(lambda x: (x-np.mean(x))/ (max(x) - min(x)))
normalised_df.head()
df['carlength'] = normalised_df['carlength']
df['carheight'] = normalised_df['carheight']
df['symboling'] = normalised_df['symboling']
df['boreratio'] = normalised_df['boreratio']
df['enginesize'] = normalised_df['enginesize']
df['stroke'] = normalised_df['stroke']
df['compressionratio'] = normalised_df['compressionratio']
df['peakrpm']= normalised_df['peakrpm']
df['horsepower'] = normalised_df['horsepower']
df['citympg'] = normalised_df['citympg']
df['price'] = normalised_df['price']
df.head()
```

## Step 9. Building Model 

```python
# Building Model after RFE
# Adding a constant variable
import statsmodels.api as sm
X_train1 = sm.add_constant(X_train1)
```

```python
# Running the linear model 1
lm_1 = sm.OLS(y_train,X_train1).fit()
print(lm_1.summary())
```

```python
# Define VIF calculation
def vif_cal(input_data, dependent_col):
    vif_df = pd.DataFrame( columns = ['field', 'vif'])
    new_df=input_data.drop([dependent_col], axis=1)
    new_dfvars=new_df.columns
    for i in range(0,new_dfvars.shape[0]):
        y=new_df[new_dfvars[i]] 
        x=new_df[new_dfvars.drop(new_dfvars[i])]
        rsq=sm.OLS(y,x).fit().rsquared  
        vif=round(1/(1-rsq),2)
        vif_df.loc[i] = [new_dfvars[i], vif]
    return vif_df.sort_values(by = 'vif', axis=0, ascending=False, inplace=False)
```
